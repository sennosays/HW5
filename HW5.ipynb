```julia
srand(42); 
A = rand(Float64,8,8); 
b = rand(Float64,8); 

x_Julia = Julia_multiply(A,b); 
x_cols = inner_cols_multiply(A,b); 
x_rows = inner_rows_multiply(A,b); 

check_equal(x_Julia,x_cols); 
check_equal(x_Julia,x_rows); 
```

I believe that for large arrays the Julia version of matrix multiply will perform the best because it allows the compiler to have the most freedom. However, I believe that the "vectorized" versions of matrix multiply may perform better for smaller matrices because there will be less overhead associated with potentially trying to change the code to optimize any loops. Finally I expect the function that places the loop over the columns index to perform better than the inner rows because of the way that arrays are stored in Julia. However, I may have this last part backwards. 

```julia
(array, Julia_time, cols_time, rows_time) = evaluate_multiply_time(10);
plot(array,Julia_time,"-o");
plot(array,cols_time,"-o"); 
plot(array,rows_time,"-o"); 
legend(loc = 2,["Julia Multplier","Column Inner Loop","Row Inner Loop"]);
title("Large Scale Performance"); 
xlabel("Array Size");
ylabel("Compute Time");
```

```julia
(array, Julia_time, cols_time, rows_time) = evaluate_multiply_time(6);
plot(array,Julia_time,"-o");
plot(array,cols_time,"-o"); 
plot(array,rows_time,"-o"); 
legend(loc = 2,["Julia Multplier","Column Inner Loop","Row Inner Loop"]);
title("Small Scale Performance");
xlabel("Array Size");
ylabel("Compute Time");
```

For small matrices of size 8 the row inner loop out performs both the defalt julia multiply function and columns inner loop. I believe this is due to the overhead of Julia optimizing matrix multiply instead of expressing it in "vectorized" form. It appears the the column inner loop always out performs the row inner loop. I believe this is due to the way a two dimential array is stored in memory, with all the elements of a column located close to each other in memory, and elements that are located along the same row offset in memory by the amount of memory for one column. It appears that with simple operations like matrix multiply it is best to let the complier have a lot of freedom, and not hard code in a particular algorithm.
